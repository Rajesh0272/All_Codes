# Now lets explore banglore property prices dataset

```python
df=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/FeatureEngineering/1_outliers/bhp.csv')
df.head()
```

```python
df.describe()
```

```python
min_2,max_2 = df['price_per_sqft'].quantile([0.001,0.999])
```

```python
df[df['price_per_sqft']<min_2]
```

```python
df[df['price_per_sqft']>max_2]
```

removal of outliers 

```python
df2=df[(df['price_per_sqft']>min_2) & (df['price_per_sqft']<max_2)]
```

```python
df2
```

```python
df2.describe()
```

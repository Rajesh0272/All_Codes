```python
corr=[]
file = open('corrupted_tokens.txt','r')
for line in file:
    x=line.strip()
    corr.append(x)
vol = []
file1 = open('vocabulary_list.txt','r')
for line in file1:
    k=line.strip()
    vol.append(k)
print('corr 1st 10 words ',corr[:10])
print('vol 1st 10 words ',vol[:10])
```

```python
import re
d=dict()
file=open('recovered_tokens.txt','a+')
file.write('\nrecovered_tokens dictionary ')
for i in corr:
    i=i.lower()
    k=[]
    for j in vol:
        j=j.lower()
        x=i.replace('#','\w')
        x='^'+x+'$'
        a=re.search(x,j)
        if a:
            k.append(a.group(0))
        else:
            pass
    
    tokens=list(set(k))
    file.write(f'\n{i} : {tokens}')
    d[i]=tokens
print(d) 
file.close()
```

Case Study 5:-
>pdb|5L4O|A Chain A, tRNA (76-MER) NGCGGGGNGGAGCAGCCTGGNAGCTCGTCGGGNTCATAACCCGAAGATCGTCGGNNCAAATCCGGCCCCCGCTACCA. Create a file (dna.txt)
Find all occurrences (single -N and double - NN) in this sting. Give the output in the file, out.txt(hint-usere)

```python
import re
string = "pdb|5L4O|A Chain A, tRNA (76-MER) NGCGGGGNGGAGCAGCCTGGNAGCTCGTCGGGNTCATAACCCGAAGATCGTCGGNNCAAATCCGGCCCCCGCTACCA"
file=open('dna.txt','a+')
file.write(f'{string}')
file2=open('output.txt','a+')   

l=['N','NN']

for i in file.readlines():
    double=re.findall(l[1],str(i))
    mixed =re.findall(l[0],str(i))
    sin = mixed[0:-len(double)*2]
file2.write(f" \nThe total no.of N in the text are : {len(mixed)}  {mixed}")  
file2.write(f" \nThe total no.of single-N in the text are : {len(sin)}  {sin}")
file2.write(f" \nThe total no.of double-N in the text are : {len(double)}  {double}")
file2.seek(0)
x=file2.read()
print(x)
file.seek(0)
file.close()
file2.close()
```

```python
# Example of Generator function
def add(mymax):
    a,b=0,1
    while True:
        c=a+b
        if c<mymax:
            yield c
            a=b
            b=c
        else:
            break
s=add(4)
```

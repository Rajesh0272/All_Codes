```python
import numpy as np 
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns  
import pandas as pd
%matplotlib inline
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/codebasics/deep-learning-keras-tf-tutorial/master/6_gradient_descent/insurance_data.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('bought_insurance',axis=1),
                                                        df['bought_insurance'],
                                                        test_size=0.2,
                                                        random_state = 25)
```

```python
x_train_scaled = x_train.copy()
x_train_scaled = x_train_scaled['age']/100
x_test_scaled = x_test.copy()
x_test_scaled = x_test_scaled['age']/100
```

```python
from tensorflow import keras
```

```python

model = keras.Sequential([
    keras.layers.Dense(1 , input_shape=(2,) ,activation = 'sigmoid',kernel_initializer='ones', bias_initializer='zeros')
])

model.compile(optimizer = 'adam' , 
             loss = 'binary_crossentropy' ,
             metrics = ['accuracy'])
model.fit(x_train_scaled , y_train , epochs=10) 
# model.fit(X_train_scaled, y_train, epochs=5000)
```
